2)	Implement the Image classification CNN model for classifying hand-written MNIST dataset by dividing the model into following 4 stages:
a.	Loading and preprocessing the image data
b.	Defining the model's architecture
c.	Training the model
d.	Estimating the model's performance


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.utils import to_categorical #used for labels, as 0 - 9 category, for each num 
```

```python
mnist_train = pd.read_csv('./datasets/mnist_train.csv')
x_train = mnist_train.drop('label',axis=1).values 
y_train = mnist_train['label'].values
#axis=1 = column, we get actual list of values, label is dropped as we want to seperate label and features

print(x_train.shape)
print(y_train.shape)
#x is features, y is labels
```

```python
mnist_test = pd.read_csv('./datasets/mnist_test.csv')
x_test = mnist_test.drop('label',axis=1).values
y_test = mnist_test['label'].values

print(x_test.shape)
print(y_test.shape)
```

```python
# Normalize pixel values to the range [0, 1]
x_train = x_train / 255.0
x_test = x_test / 255.0

#flatten - means we want to convert it to 1D array instead of its 2D form
x_train = x_train.reshape((-1, 28, 28, 1))
x_test = x_test.reshape((-1, 28, 28, 1))
#as the image was already flattened when we import the data from .csv file, 
#so we will convert it to 2D form using reshape, -1 is auto value, 
#1 is the depth dimension/single dimesnion, if it was RGB it would hve been 3

print(x_train.shape)
print(x_test.shape)
```

```python
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
#to_categoricl convers the values into a vector wehere the val is set to 1 if the digit matches in vector
```

```python
#CNN

model = Sequential([ #this allows u to write ur layes in the form of a list
    Input(shape=(28,28,1)), #this val is same as x_train.shape
    Conv2D(32,(3,3),activation='relu',input_shape=(28,28,1)),
    #It performs convolution operations to extract spatial features from the input image.
    #Spatial features refer to patterns or structures in the image that represent important information.
    
    MaxPooling2D((2,2)),
    #This is a pooling layer that performs down-sampling of the input by taking the maximum value from a (2, 2) window.
    
    Conv2D(64,(3,3),activation='relu'),
    MaxPooling2D((2,2)),
    Flatten(),
    Dense(64,activation='relu'),
    Dense(10,activation='softmax')
])
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
#adam is genreally used as its applicable for most of the tasks and its robust
#sparse_categorical_crossentropy - This loss function is used when the target labels are integers, representing the class indices directly. 
#categorical_corssentropy - This loss function is used when the target labels are one-hot encoded vectors.
```

```python
history = model.fit(x_train,y_train,epochs=10,validation_data=(x_test,y_test))
```

```python
loss, accuracy = model.evaluate(x_test,y_test)
```

```python
plt.plot(history.history['loss'],label='Training Loss')
plt.plot(history.history['val_loss'],label='Validation Loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend()
plt.show()
```

```python
plt.plot(history.history['accuracy'],label='Training Accuracy')
plt.plot(history.history['val_accuracy'],label='Validation Accuracy')
plt.xlabel('epochs')
plt.ylabel('ccuracy')
plt.legend()
plt.show()
```

```python
y_pred = model.predict(x_test) # to predict we need features so we give x_test
```
